```python
import numpy as np

def linear_regression(x, y):
    n = len(x)
    sum_x = np.sum(x)
    sum_y = np.sum(y)
    sum_x_squared = np.sum(x**2)
    sum_xy = np.sum(x * y)

    a = (sum_y * sum_x_squared - sum_x * sum_xy) / (n * sum_x_squared - sum_x**2)
    b = (n * sum_xy - sum_x * sum_y) / (n * sum_x_squared - sum_x**2)

    return a, b

def sigma_r_squared(y, y_pred, p):
    n = len(y)
    residual_variance = np.sum((y - y_pred)**2) / (n - p)
    return residual_variance

def R_squared(y, y_pred, sigma_y_squared, p):
    n = len(y)
    sigma_r_squared_val = sigma_r_squared(y, y_pred, p)
    R_squared_val = 1 - ((n - p) * sigma_r_squared_val) / ((n - 1) * sigma_y_squared)
    return R_squared_val



np.random.seed(0)
x1 = np.random.rand(100)
y1 = np.random.rand(100)

x2 = np.random.rand(100)
y2 = 2 * x2 + 0.5 * np.random.randn(100)

x3 = np.random.rand(100)
y3 = np.sin(3 * np.pi * (x3 + 0.8)) + 2 * x3 + 0.2 * np.random.randn(100)

a1, b1 = linear_regression(x1, y1)
a2, b2 = linear_regression(x2, y2)
a3, b3 = linear_regression(x3, y3)

sigma_y1_squared = np.std(y1)**2
sigma_y2_squared = np.std(y2)**2
sigma_y3_squared = np.std(y3)**2

y_pred1 = a1 + b1 * x1
y_pred2 = a2 + b2 * x2
y_pred3 = a3 + b3 * x3

sigma_r_squared1 = sigma_r_squared(y1, y_pred1, p=2)
R_squared1 = R_squared(y1, y_pred1, sigma_y1_squared, p=2)

sigma_r_squared2 = sigma_r_squared(y2, y_pred2, p=2)
R_squared2 = R_squared(y2, y_pred2, sigma_y2_squared, p=2)

sigma_r_squared3 = sigma_r_squared(y3, y_pred3, p=2)
R_squared3 = R_squared(y3, y_pred3, sigma_y3_squared, p=2)

print("Bondad del ajuste para el conjunto de datos 1 (R^2):", R_squared1)
print("Bondad del ajuste para el conjunto de datos 2 (R^2):", R_squared2)
print("Bondad del ajuste para el conjunto de datos 3 (R^2):", R_squared3)
```

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
x3 = np.random.rand(100)
y3 = np.sin(3 * np.pi * (x3 + 0.8)) + 2 * x3 + 0.2 * np.random.randn(100)

def modelo(x, a, b, c):
    return np.sin(a * (x + b)) + c * x

def derivada_modelo(x, a, b, c):
    d_a = x * np.cos(a * (x + b))
    d_b = a * np.cos(a * (x + b))
    d_c = x
    return np.array([d_a, d_b, d_c])

def evaluar_modelo(x, params):
    a, b, c = params
    return modelo(x, a, b, c), derivada_modelo(x, a, b, c)

def newton_raphson(x, y, params_init, max_iter=100, tol=1e-6):
    params = params_init
    for _ in range(max_iter):
        y_pred, d_modelo = evaluar_modelo(x, params)
        r = y - y_pred
        J = d_modelo.T
        JtJ = np.dot(J.T, J)
        Jtr = np.dot(J.T, r)
        try:
            delta_params = np.linalg.solve(JtJ, Jtr)
        except np.linalg.LinAlgError:
            print("Singular matrix encountered. Optimization cannot proceed.")
            break
        params += delta_params
        if np.linalg.norm(delta_params) < tol:
            break
    return params



#params_init = np.array([9.496806581662378, -0.5307754979326218, 2.0187681244165003, 0.031050703479117616])

params_init = np.array([7., 0., 0.])


a_opt, b_opt, c_opt = newton_raphson(x3, y3, params_init)

x_fit = np.linspace(0, 1, 100)
y_fit = modelo(x_fit, a_opt, b_opt, c_opt)


plt.scatter(x3, y3, label='Datos originales')
plt.plot(x_fit, y_fit, color='red', label='Curva ajustada')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Ajuste de curva con Newton-Raphson')
plt.legend()
plt.show()
```

```python
def newton_raphson_silent(x, y, params_init, max_iter=100, tol=1e-6):
    params = params_init
    for _ in range(max_iter):
        y_pred, d_modelo = evaluar_modelo(x, params)
        r = y - y_pred
        J = d_modelo.T
        JtJ = np.dot(J.T, J)
        Jtr = np.dot(J.T, r)
        try:
            delta_params = np.linalg.solve(JtJ, Jtr)
        except np.linalg.LinAlgError:
            break
        params += delta_params
        if np.linalg.norm(delta_params) < tol:
            break
    return params

params_init1 = np.array([7., 0., 0.])

params_init2 = np.array([5, 0., 0.])

a_opt1, b_opt1, c_opt1 = newton_raphson_silent(x3, y3, params_init1)

a_opt2, b_opt2, c_opt2 = newton_raphson_silent(x3, y3, params_init2)

y_fit1 = modelo(x_fit, a_opt1, b_opt1, c_opt1)
y_fit2 = modelo(x_fit, a_opt2, b_opt2, c_opt2)

plt.figure(figsize=(14, 7))

plt.subplot(1, 2, 1)
plt.scatter(x3, y3, label='Datos originales')
plt.plot(x_fit, y_fit1, color='red', label='Curva ajustada (params_init1)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Ajuste con params_init1')
plt.legend()

plt.subplot(1, 2, 2)
plt.scatter(x3, y3, label='Datos originales')
plt.plot(x_fit, y_fit2, color='blue', label='Curva ajustada (params_init2)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Ajuste con params_init2')
plt.legend()

plt.tight_layout()
plt.show()
```

Implementación del método de Newton-Raphson para optimizar los parámetros de un modelo matemático basado en datos. Este método es especialmente útil para encontrar las raíces de una función o para ajustar modelos no lineales a datos experimentales o teóricos.

$x_{k+1} = x_k - \frac{F(x_k)}{F'(x_k)}$

es la fórmula iterativa del método de Newton-Raphson para encontrar las raíces de una función real $F(x)$. Esta ecuación describe cómo actualizar el valor de $x$ en cada paso $s$ para acercarse a una raíz de $F$, es decir, un valor de $x$ para el cual $F(x) = 0$.

### Generalización para Múltiples Dimensiones
Consideremos el caso de un ajuste donde buscamos los parámetros óptimos que ajustan un modelo a unos datos. Por tanto, sustituimos las $x$ por $p$ y generalizamos a múltiples dimensiones:

$\mathbf{p}_{k+1} = \mathbf{p}_k - \mathbf{H}^{-1} \nabla f(\mathbf{p})$

donde $\mathbf{H}$ es la matriz Hessiana.

Dada una función $f: \mathbb{R}^n \rightarrow \mathbb{R}$ que depende de $n$ variables, $p_1, p_2, \ldots, p_n$, la matriz Hessiana $H$ de $f$ se define como:


$H(f) =
\begin{bmatrix}
\frac{\partial^2 f}{\partial p_1^2} & \frac{\partial^2 f}{\partial p_1 \partial p_2} & \cdots & \frac{\partial^2 f}{\partial p_1 \partial p_n} \\
\frac{\partial^2 f}{\partial p_2 \partial p_1} & \frac{\partial^2 f}{\partial p_2^2} & \cdots & \frac{\partial^2 f}{\partial p_2 \partial p_n} \\
\vdots  & \vdots  & \ddots & \vdots  \\
\frac{\partial^2 f}{\partial p_n \partial p_1} & \frac{\partial^2 f}{\partial p_n \partial p_2} & \cdots & \frac{\partial^2 f}{\partial p_n^2}
\end{bmatrix}$


Cada elemento $H_{ij}$ de la matriz Hessiana es la segunda derivada parcial de $f$ con respecto a $p_i$ y $p_j$:

$H_{ij} = \frac{\partial^2 f}{\partial p_i \partial p_j}$


El método se generaliza a múltiples dimensiones al considerar $F(x)$ como un vector de residuos (diferencias entre los datos observados y las predicciones del modelo) y $F'(x)$ como la matriz Jacobiana $J$ de derivadas parciales de los residuos con respecto a los parámetros del modelo. La actualización de los parámetros del modelo, en este caso, sigue el esquema:

$\mathbf{p}_{k+1} = \mathbf{p}_k - (J^TJ)^{-1}J^T\mathbf{r}$

Aquí, $\mathbf{p}_k$ es el vector de parámetros en la iteración $k$, $J$ es la matriz Jacobiana de las derivadas de los residuos con respecto a los parámetros, y $\mathbf{r}$ es el vector de residuos. La expresión $(J^TJ)^{-1}J^T\mathbf{r}$ es el equivalente multidimensional del cociente $\frac{F(x_s)}{F'(x_s)}$ en la ecuación unidimensional.


Aquí, $J$ es la matriz Jacobiana de los residuos con respecto a los parámetros $\mathbf{p}$, y $(J^TJ)$ actúa como una aproximación de la matriz Hessiana para el problema de minimización de los cuadrados de los residuos. Esto se debe a que, en el caso de los mínimos cuadrados, la matriz $J^TJ$ representa efectivamente la curvatura de la función de coste cuadrático respecto a los parámetros.


### Interpretación
- **Unidimensional (Búsqueda de Raíces):** La fórmula original busca iterativamente una raíz de la función \(F(x)\), usando la derivada \(F'(x)\) para ajustar la dirección y magnitud del paso.
- **Multidimensional (Optimización de Parámetros):** En optimización, se busca el conjunto de parámetros que minimiza la discrepancia (residuos) entre las predicciones del modelo y los datos observados. La matriz Jacobiana y su transpuesta reemplazan a la derivada en la versión unidimensional, proporcionando la dirección y tamaño del paso en el espacio de parámetros para mejorar el ajuste del modelo a los datos.

En ambos casos, el método se basa en la idea de usar la información de derivadas (o gradientes) para hacer ajustes iterativos en busca de un objetivo: encontrar una raíz o minimizar una función de costo. La eficacia del método depende de la selección adecuada de los valores iniciales y de las propiedades de la función objetivo, como su continuidad y la existencia de derivadas.

### Inicialización
- **`params_init`**: Vector de parámetros iniciales $\mathbf{p}_0 = [p_{1,0}, p_{2,0}, \ldots, p_{n,0}]$.
- **`max_iter`**: Número máximo de iteraciones permitidas.
- **`tol`**: Tolerancia para la norma del vector de cambio de parámetros $\Delta\mathbf{p}$, usada como criterio de convergencia.

### Proceso Iterativo
Para cada iteración $k$, el método realiza los siguientes pasos:

1. **Evaluación del Modelo y su Derivada**:
   - Calcula las predicciones del modelo $y_{\text{pred}}$ y la matriz de derivadas (Jacobiano, $J$) basado en los parámetros actuales $\mathbf{p}_k$.

2. **Cálculo de los Residuos**:
   - Los residuos son la diferencia entre los datos observados $y$ y las predicciones del modelo $y_{\text{pred}}$:

     $ \mathbf{r} = \mathbf{y} - \mathbf{y}_{\text{pred}} $

3. **Formación del Sistema de Ecuaciones Lineales**:
   - Construye la matriz $J^TJ$ y el vector $J^T\mathbf{r}$ para formar el sistema de ecuaciones lineales:
   
     $ J^TJ \Delta\mathbf{p} = J^T\mathbf{r} $
   - Aquí, $J^T$ es la transpuesta de la matriz Jacobiana.

4. **Solución del Sistema de Ecuaciones**:
   - Resuelve el sistema lineal para encontrar $\Delta\mathbf{p}$, el cambio en los parámetros:
     
     $ \Delta\mathbf{p} = (J^TJ)^{-1}J^T\mathbf{r} $
   - En la práctica, se utiliza `np.linalg.solve` para obtener $\Delta\mathbf{p}$ de manera más numéricamente estable que calcular explícitamente la inversa.

5. **Actualización de Parámetros**:
   - Actualiza los parámetros usando $\Delta\mathbf{p}$:

     $ \mathbf{p}_{k+1} = \mathbf{p}_k - \Delta\mathbf{p} $

6. **Criterio de Convergencia**:
   - Si la norma de $\Delta\mathbf{p}$ es menor que la tolerancia especificada (`tol`), el proceso iterativo se detiene, indicando que se ha alcanzado la convergencia.

### Devolución de Resultados
- **`params`**: Vector de parámetros optimizados $\mathbf{p}_{\text{opt}}$.

Este método iterativo se repite hasta que se cumple el criterio de convergencia o se alcanza el número máximo de iteraciones. Es importante notar que el éxito y la velocidad de convergencia del método de Newton-Raphson pueden depender significativamente de los valores iniciales de los parámetros $\mathbf{p}_0$.

```python
def calcular_BIC(n, p, sigma_r_squared):
    """
    Calcula el Criterio de Información Bayesiano (BIC).

    Parámetros:
    - n (int): Número de puntos de datos.
    - p (int): Número de parámetros libres en el modelo.
    - sigma_r_squared (float): Varianza residual de los errores del modelo.

    Retorna:
    - BIC (float): Valor del Criterio de Información Bayesiano.
    """
    BIC = n * np.log((n-p)/n*sigma_r_squared) + p * np.log(n)
    return BIC
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

np.random.seed(0)
x3 = np.random.rand(100)

# Generar y3 con errores que dependen de x3
errores = 0.2 * (1 + x3)**2  # Los errores aumentan con x3
y3 = np.sin(3 * np.pi * (x3 + 0.8)) + 2 * x3 + errores * np.random.randn(100)

def modelo(x, a, b, c, d):
    return np.sin(a * (x + b)) + c * x + d

# Utilizar los errores como pesos inversos para el ajuste
popt, pcov = curve_fit(modelo, x3, y3, sigma=errores, absolute_sigma=True, maxfev=5000)
#popt, pcov = curve_fit(modelo, x3, y3, maxfev=5000)

a_opt, b_opt, c_opt, d_opt = popt

print(a_opt,b_opt,c_opt,d_opt)

# Generar datos para la curva ajustada
x_fit = np.linspace(0, 1, 100)
y_fit = modelo(x_fit, a_opt, b_opt, c_opt, d_opt)

# Visualización de los datos y la curva ajustada
plt.scatter(x3, y3, label='Datos originales')
plt.errorbar(x3, y3, yerr=errores, fmt='o', color='gray', alpha=0.5, label='Errores')
plt.plot(x_fit, y_fit, color='red', label='Curva ajustada')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Ajuste Mejorado con Tendencia Sinusoidal y Lineal')
plt.legend()
plt.show()
```
